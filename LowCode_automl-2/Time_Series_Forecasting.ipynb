*PyCaret's time series forecasting module is now available. The module currently is suitable for univariate / multivariate time series forecasting tasks. The API of time series module is consistent with other modules of PyCaret.It comes built-in with preprocessing capabilities and over 30 algorithms comprising of statistical / time-series methods as well as machine learning based models.*

```python
### loading sample dataset from pycaret dataset module
import pandas as pd
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv'
data = pd.read_csv(url, header=0, index_col=0, parse_dates=True)
```

```python
# plot the dataset
data.plot()
```

```python
# import pycaret time series and init setup
from pycaret.time_series import *
s = setup(data,fh = 3, session_id = 123)
```

```python
# import TSForecastingExperiment and init the class
from pycaret.time_series import TSForecastingExperiment
exp = TSForecastingExperiment()
```

```python
type(exp)
```

```python
# init setup on exp
exp.setup(data, fh = 3, session_id = 123)
```

```python
# check statistical tests on original data
check_stats()
```

```python
# compare baseline models
best = compare_models()
```

```python
# compare models using OOP
exp.compare_models()
```

```python
# plot forecast
plot_model(best, plot = 'forecast')
```

```python
# plot forecast for 36 months in future
plot_model(best, plot = 'forecast', data_kwargs = {'fh' : 36})
```

```python
# residuals plot
plot_model(best, plot = 'residuals')
```

```python
# predict on test set
holdout_pred = predict_model(best)
```

```python
# show predictions df
holdout_pred.head()
```

```python
# generate forecast for 36 period in future
predict_model(best, fh = 36)
```

```python
# save pipeline
save_model(best, 'my_first_pipeline')
```
