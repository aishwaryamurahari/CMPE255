*PyCaret's Clustering Module is an unsupervised machine learning module that performs the task of grouping a set of objects in such a way that objects in the same group (also known as a cluster) are more similar to each other than to those in other groups.*

*A typical workflow in PyCaret's unsupervised module consist of following 6 steps in this order:*
Setup, create model, assign labels, analyze model, prediction and save model

```python
from pycaret.datasets import get_data
data = get_data('credit')
```

*This function initializes the training environment and creates the transformation pipeline. Setup function must be called before executing any other function in PyCaret. It only has one required parameter. Session_id is a pseudo-random number distributed as a seed in all functions for later reproducibility. If no session_id is passed, a random number is automatically generated that is distributed to all functions.*

```python
from pycaret.clustering import *
s = setup(data, session_id = 123)
```

*Import ClusteringExperiment and init the class*

```python
from pycaret.clustering import ClusteringExperiment
exp = ClusteringExperiment()
```

```python
type(exp)
```

*init setup on exp*

```python
exp.setup(data, session_id = 123)
```

**Create Model** :  *This function trains and evaluates the performance of a given model. Metrics evaluated can be accessed using the get_metrics function. Custom metrics can be added or removed using the add_metric and remove_metric function.*

*Train Kmeans model*

```python
kmeans = create_model('kmeans')
```

*To check all the available models*

```python
models()
```

*Train meanshift model*

```python
meanshift = create_model('meanshift')
```

**Assign Model**: *This function assigns cluster labels to the training data, given a trained model.*

```python
kmeans_cluster = assign_model(kmeans)
kmeans_cluster
```

**Analyze Model:** *You can use the plot_model function to analyzes the performance of a trained model on the test set. It may require re-training the model in certain cases.*

*Plot pca cluster plot*

```python
plot_model(kmeans, plot = 'cluster')
```

*Plot elbow plot*

```python
plot_model(kmeans, plot = 'elbow')
```

*Plot silhouette*

```python
plot_model(kmeans, plot = 'silhouette')
```

*An alternate to plot_model function is evaluate_model.*

```python
evaluate_model(kmeans)
```

Prediction: *The predict_model function returns Cluster label as a new column in the input dataframe. predict_model is only useful when you want to obtain cluster labels on unseen data (i.e. data that was not used during training the model).*

*Predict on test set*

```python
kmeans_pred = predict_model(kmeans, data=data)
kmeans_pred
```

**Save Model:** *Finally, you can save the entire pipeline on disk for later use, using pycaret's save_model function.*

```python
save_model(kmeans, 'kmeans_pipeline')
```

*Load pipeline*

```python
kmeans_pipeline = load_model('kmeans_pipeline')
kmeans_pipeline
```
